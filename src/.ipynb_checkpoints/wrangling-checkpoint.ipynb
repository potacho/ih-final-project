# Data Wrangling

Includes:

- More specific cleaning (deleting all rows before power sensors were installed and columns with un-interpretable sensors).

- Input dataframe for every analysis.


```python
import pandas as pd
import numpy as np
import re
```

```python
df_01 = pd.read_csv('./output/data_01.csv',sep=',',index_col=0)
```

```python
df_02 = df_01.copy()
```

```python
df_02.loc[df_02['DateTime'] == '2010-05-11 00:00:00']
```

```python
df_02.drop(df_02.index[0:5904],inplace=True)
df_02 = df_02.reset_index(drop=True)
```

```python
df_02 = df_02.drop(['T001','M008','M017','M019','T003'], axis=1)
```

```python
display(df_02.head(),df_02.tail(),df_02.shape)
```

```python
df_02.to_csv(r'./output/data_02.csv')
```

### Dataframe for Behavior Explanatory Models (BEM)

```python
df_bem = df_02.copy()
```

```python
def time_constrictor3(tiempo_string):
    tiempo_string = re.sub(r' \d\d:\d\d:\d\d$','',tiempo_string)
    return tiempo_string
```

```python
df_bem['DateTime'] = df_bem['DateTime'].apply(time_constrictor3)
```

```python
df_bem['DateTime'] = pd.to_datetime(df_bem['DateTime'])
```

```python
df_bem = df_bem.groupby(['DateTime'],as_index=False).mean()
```

```python
df_bem.to_csv(r'./output/df_bem.csv')
```

### Dataframe for Sensor Profile Comparison (SP)

```python
df_sp = df_bem.copy()
```

```python
df_sp = df_sp.loc[df_sp['DayWeek'] == 5]
```

```python
df_sp.to_csv(r'./output/df_sp.csv')
```
